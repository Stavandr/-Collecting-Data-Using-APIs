# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
conn = sqlite3.connect("m4_survey_data.sqlite")

query = "SELECT * FROM master;"

df = pd.read_sql(query, conn)

df.plot(kind='hist', x='Age', y='WorkWeekHrs', figsize=(10, 6), color='darkblue')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT * FROM master
"""
df = pd.read_sql_query(QUERY,conn)
df.boxplot(column='Age')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

QUERY = """
SELECT * FROM master
"""
df = pd.read_sql_query(QUERY,conn)
plot = sns.scatterplot(x='Age', y='WorkWeekHrs', data=df)
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
"""

# Execute the query and load the data into a DataFrame
df1 = pd.read_sql_query(QUERY, conn)

# Plot scatter plot using Matplotlib
plt.scatter(df1['WorkWeekHrs'], df1['CodeRevHrs'], s=df1['Age'], c=df1['Age'], alpha=0.5)
plt.title('WorkWeekHrs and CodeRevHrs By Age', size=14)
plt.xlabel('WorkWeekHrs', size=10)
plt.ylabel('CodeRevHrs', size=10)
plt.colorbar(label='Age')
plt.show() 
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
select count(DatabaseDesireNextYear) as Count, DatabaseDesireNextYear from DatabaseDesireNextYear group by DatabaseDesireNextYear order by count(DatabaseDesireNextYear) DESC LIMIT 5
"""

df2=pd.read_sql_query(QUERY,conn)
df2.set_index('DatabaseDesireNextYear',inplace=True)
df2

labels = df2['Count']
sizes = df2['Count']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
print(df2)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age 
FROM master 
WHERE Age BETWEEN 30 AND 35
"""

# Execute the query and load the data into a DataFrame
df3 = pd.read_sql_query(QUERY, conn)

# Group by Age and calculate median
df_median = df3.groupby(df3.Age)[['WorkWeekHrs', 'CodeRevHrs']].median()

# Reset index for plotting
df_median.reset_index(inplace=True)

# Melt the DataFrame to have 'Age' as the id variable
df_melted = df_median.melt(id_vars='Age', var_name='Metrics', value_name='Median Hours')

# Plotting using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=df_melted, x='Age', y='Median Hours', hue='Metrics', palette='viridis', edgecolor='w')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp FROM master where Age BETWEEN 45 and 60
"""

df4 = pd.read_sql_query(QUERY, conn)
df4.head()

df4.plot(kind='line')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(*) AS MainBranch_Count
FROM master
GROUP BY MainBranch
"""

# Execute the query and load the data into a DataFrame
df5 = pd.read_sql_query(QUERY, conn)

# Plotting using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=df5, x='MainBranch_Count', y='MainBranch', palette='viridis')
plt.title('Counts of MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.show()
```
